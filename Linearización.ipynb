#Linearización y equilibrios no lineales

#Linearización de funciones multivariadas

Sea $f(x,y):\mathbb{R^2\rightarrow\mathbb{R}}$ una función real multivariada y asumamos que $x$ y $y$ ambas dependen de $t$. Entonces $g(t)=f(x(t),y(t))$ tiene derivada $$g'(t)=f_x(x(t),y(t))x'(t)+f_y(x(t),y(t))y'(t),$$ como consecuencia de la regla de la cadena. Por tanto, la linearización de $g(t)$ alrededor de $t_0$ es $$g(t)\approx g(t_0)+g'(t_0)(t-t_0)=f(x(t_0),y(t_0))+f_x(x(t_0),y(t_0))x'(t_0)(t-t_0)+f_y(x(t_0),y(t_0))y'(t_0)(t-t_0).$$

Es decir,
$$g(t)=f(x,y)\approx f(x(t_0),y(t_0))+f_x(x(t_0),y(t_0))\frac{dx}{dt}\Delta t+f_y(x(t_0),y(t_0))\frac{dx}{dt}\Delta t,$$ o
$$f(x,y)\approx f(x_0,y_0)+f_x(x_0,y_0)\Delta x+f_y(x_0,y_0)\Delta y\approx f(x_0,y_0)+f_x(x_0,y_0)(x-x_0)+f_y(x_0,y_0)(y-y_0).$$

# Linearización de funciones vectoriales

Tomemos una función vectorial $$\overline{h}((x,y))=\begin{pmatrix}f(x,y) \\ g(x,y) \end{pmatrix}$$ y encontremos su linearización alrededor de $(x_0,y_0)$:

$$\overline{h}(x,y)≈\begin{pmatrix}f(x_0,y_0) \\ g(x_0,y_0) \end{pmatrix}+\begin{pmatrix}f_x(x_0,y_0) & f_y(x_0,y_0) \\ g_x(x_0,y_0) & g_y(x_0,y_0) \end{pmatrix}\begin{pmatrix}x-x_0 \\ y-y_0 \end{pmatrix}$$

Esa matriz de derivadas parciales de las componentes de la función vectorial se llama el *Jacobiano* asociado a la función.

# Sistemas de EDO's autónomos

Consideraremos un sistema de EDOs autónomo: $$\vec{x}'=\vec{F}(\vec{x}).$$ En particular, estaremos trabajando con el caso bi-dimensional: $$\begin{pmatrix} x(t)\\y(t) \end{pmatrix}'=\begin{pmatrix} f(x(t),y(t))\\g(x(t),y(t)) \end{pmatrix}.$$

Buscaremos *puntos de equilibrio*, es decir, puntos en el plano $(x,y)$ donde $$\begin{pmatrix} x(t)\\y(t) \end{pmatrix}'=\begin{pmatrix} f(x(t),y(t))\\g(x(t),y(t)) \end{pmatrix}=\begin{pmatrix} 0\\0 \end{pmatrix}.$$

Entonces, linearizaremos el sistema alrededor de los puntos de equilibrio. Estudiaremos la naturaleza (estabilidad) de los sistemas linearizados y extrapolaremos al sistema no lineal original.

#Sistemas de EDOs linearizados alrededor de equilibrios

Trabajando con el caso bi-dimensional como ejemplo, buscamos equilibrios: $$\begin{pmatrix} x\\y \end{pmatrix}'=\begin{pmatrix} f(x,y)\\g(x,y) \end{pmatrix}=\begin{pmatrix} 0\\0 \end{pmatrix}.$$
Llamemos $(x_0,y_0)$ a uno de los equilibrios y linearicemos:$$\begin{pmatrix} x\\y \end{pmatrix}'=\begin{pmatrix} f(x,y)\\g(x,y) \end{pmatrix}\approx\begin{pmatrix} f(x_0,y_0)\\g(x_0,y_0) \end{pmatrix}+\begin{pmatrix}f_x(x_0,y_0) & f_y(x_0,y_0) \\ g_x(x_0,y_0) & g_y(x_0,y_0) \end{pmatrix}\begin{pmatrix}x-x_0 \\ y-y_0 \end{pmatrix}.$$
Recordando que $(x_0,y_0)$ es un equilibrio y por tanto $$\begin{pmatrix}f(x_0,y_0) \\ g(x_0,y_0) \end{pmatrix}=\begin{pmatrix}0 \\ 0 \end{pmatrix},$$
tenemos entonces $$\begin{pmatrix} x-x_0\\y-y_0 \end{pmatrix}'=\begin{pmatrix}f_x(x_0,y_0) & f_y(x_0,y_0) \\ g_x(x_0,y_0) & g_y(x_0,y_0) \end{pmatrix}\begin{pmatrix}x-x_0 \\ y-y_0 \end{pmatrix},$$
un sistema lineal autónomo.

# Relación entre ecuaciones de orden superior y sistemas de EDOs de 1er orden: el péndulo no lineal

Tomemos la ecuación no lineal del péndulo, de masa $m$ y longitud $l$, donde el ángulo con respecto a la vertical es $\theta$: $$ml\theta''=-mg\sin(\theta)$$
image.png

Por tanto, obtenemos $$\theta ''=-\frac{g}{l}\sin(\theta).$$

Si definimos $x=\theta$ y $y=x'=\theta'$, entonces podemos escribir esta ecuación de 2do orden como el siguiente sistema de ecuaciones de 1er orden:
$$ \begin{pmatrix} x\\y\end{pmatrix}'=\begin{pmatrix} y\\-\frac{g}{l}\sin(x)\end{pmatrix},$$
donde $\theta(t)$, $x(t)$ y $y(t)$ son funciones del tiempo $t$, la variable independiente con respecto a la cual se deriva.

Similarmente podemos convertir cualquier EDO de orden superior en un sistema de EDOs de 1er orden.

#Sistemas de EDOs lineales, homogeneos y autónomos

Consideremos el siguiente sistema general de ecuaciones diferenciales lineales, homogeneas y autónomas: $$\vec{z}'=A\vec{z}.$$

Si asumimos soluciones de la forma $\vec{z(t)}=\vec{z_0}e^{\lambda t}$, sustituyendo, obtenemos: $$\lambda \vec{z_0}e^{\lambda t}=A\vec{z_0}e^{\lambda t}$$ o equivalentemente, $$A\vec{z_0}=\lambda \vec{z_0},$$ es decir, el problema de valores y vectore propios para la matriz $A$.

De esta forma, si tenemos dos valores propios reales $\lambda_1,\lambda_2$ y dos vectores propios linealmente independientes $\vec{z_1},\vec{z_2}$, tenemos dos soluciones $\vec{z_1}e^{\lambda_1 t},\vec{z_2}e^{\lambda_2 t}$.

Si los valores propios son complejos, podemos combinar las soluciones linealmente y usar el teorema de Euler ($e^{i\theta}=\cos(\theta)+i\sin(\theta)$) para encontrar soluciones reales.

#Clasificación de los equilibirios para el sistema del péndulo no lineal

Es es el sistema correspondiente al péndulo no lineal: $$ \begin{pmatrix} x\\y\end{pmatrix}'=\begin{pmatrix} y\\-\frac{g}{l}\sin(x)\end{pmatrix}.$$

Por inspección vemos que hay dos equilibrios representativos: $(0,0)$ y $(\pi,0)$ (se repiten cada $2\pi$ unidades en $x$).

Linearizando alrededor de $(0,0)$: $$ \begin{pmatrix} x\\y\end{pmatrix}'=\begin{pmatrix} y\\-\frac{g}{l}x\end{pmatrix}=\begin{pmatrix} 0 & 1\\-\frac{g}{l} & 0\end{pmatrix}\begin{pmatrix} x\\y\end{pmatrix}.$$
Por tanto, el polinomio característico es $\lambda^2+\frac{g}{l}$. Sus raices son $\pm i\sqrt{\frac{g}{l}}$, puramente imaginarias. Así que este equilibrio es un *centro*, una órbita periódica en el plano fase.

Linearizando alrededor de $(\pi,0)$: $$ \begin{pmatrix} x\\y\end{pmatrix}'=\begin{pmatrix} y\\-\frac{g}{l}x\end{pmatrix}=\begin{pmatrix} 0 & 1\\\frac{g}{l} & 0\end{pmatrix}\begin{pmatrix} x\\y\end{pmatrix}.$$
Por tanto, el polinomio característico es $\lambda^2-\frac{g}{l}$. Sus raices son $\pm \sqrt{\frac{g}{l}}$, reales de signo contrario. Así que este equilibrio es *inestable*, una silla de montar en el plano fase.


#Código Python para linearizar y graficar los campos vectoriales asociados a sistemas de EDOs de 1er orden.

```python
from sympy import *
#init_printing(use_latex=True)
init_printing(use_unicode=True)
#init_printing()
# Define the symbols
x,y = symbols( 'x y')
a,b = symbols( 'a b')


# Define the function we are going to linearize
f = Matrix( [y , -x] )

f_at = f.subs( [(x,a), (y,b)] )
J = f.jacobian([x,y])
J_at = J.subs( [(x,a), (y,b)] )
delta = Matrix([x,y]) - Matrix([a,b])

# Tailor Series Linearization
# f_approx = (f_at + J_at * delta).expand()
f_approx = f_at + J_at * delta
f_a=f_approx.subs( [(a,1), (b,2)] )
print(f_a)
latex(f_a)
```

```python
#Plotting vector fields
import matplotlib.pyplot as plt
import numpy as np

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))


u=y; v=-x

fig, ax = plt.subplots(figsize=(10,6))
q = ax.quiver(x, y, u, v)
ax.set_title(r'Vector field $\vec{F} = \frac{1}{8}(2xy\vec{i}+y^2\vec{j})$')
ax.quiverkey(q, 0.8, 0.9, 1, r'unit length', labelpos='E',
                   coordinates='figure')

fig.savefig('vectorfield2.png')
plt.show()
```
